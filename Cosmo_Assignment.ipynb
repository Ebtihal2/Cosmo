```python
from classy import Class
import matplotlib.pyplot as plt
import numpy as np
from math import pi
```

```python
# create instance of the class "Class"
LambdaCDM = Class()
# pass input parameters
LambdaCDM.set({'omega_b':0.0223828,'Omega_k':0,'h':0.67810,'A_s':2.100549e-09,'n_s':0.9660499,'tau_reio':0.05430842})
LambdaCDM.set({'output':'tCl,pCl,lCl,mPk','lensing':'yes','P_k_max_1/Mpc':3.0})
# run class
LambdaCDM.compute()
```

```python
# get all C_l output
cls = LambdaCDM.lensed_cl(2500)
# To check the format of cls
cls.keys()

ll1 = cls['ell'][2:]
clTT1 = cls['tt'][2:]
clEE = cls['ee'][2:]
clPP = cls['pp'][2:]
```

```python
plt.figure(1)
plt.xscale('log');plt.yscale('linear');plt.xlim(2,2500);plt.ylim(10**(-17),10**(-9))
plt.xlabel(r'$\ell$')
plt.ylabel(r'$[\ell(\ell+1)/2\pi]  C_\ell^\mathrm{TT}$')
plt.plot(ll1,clTT1*ll1*(ll1+1)/2./pi,'r-')
```

```python
#Increase Omega bh by 50%
LambdaCDM1 = Class()
# pass input parameters
LambdaCDM1.set({'omega_b':0.0223828*1.5,'Omega_k':0,'h':0.67810,'A_s':2.100549e-09,'n_s':0.9660499,'tau_reio':0.05430842})
LambdaCDM1.set({'output':'tCl,pCl,lCl,mPk','lensing':'yes','P_k_max_1/Mpc':3.0})
# run class
LambdaCDM1.compute()
# get all C_l output
cls = LambdaCDM1.lensed_cl(2500)
# To check the format of cls
cls.keys()

ll4 = cls['ell'][2:]
clTT4 = cls['tt'][2:]
clEE = cls['ee'][2:]
clPP = cls['pp'][2:]
plt.figure(1)
plt.xscale('log');plt.yscale('linear');plt.xlim(2,2500);plt.ylim(10**(-17),10**(-9))
plt.xlabel(r'$\ell$')
plt.ylabel(r'$[\ell(\ell+1)/2\pi]  C_\ell^\mathrm{TT}$')
plt.plot(ll4,clTT4*ll4*(ll4+1)/2./pi,'r-')
```

```python
#Increase Omega k is 0.5
LambdaCDM1 = Class()
# pass input parameters
LambdaCDM1.set({'omega_b':0.0223828,'Omega_k':0.5,'h':0.67810,'A_s':2.100549e-09*1.1,'n_s':0.9660499,'tau_reio':0.05430842})
LambdaCDM1.set({'output':'tCl,pCl,lCl,mPk','lensing':'yes','P_k_max_1/Mpc':3.0})
# run class
LambdaCDM1.compute()
# get all C_l output
cls = LambdaCDM1.lensed_cl(2500)
# To check the format of cls
cls.keys()

ll5 = cls['ell'][2:]
clTT5 = cls['tt'][2:]
clEE = cls['ee'][2:]
clPP = cls['pp'][2:]
plt.figure(1)
plt.xscale('log');plt.yscale('linear');plt.xlim(2,2500);plt.ylim(10**(-17),0.2*10**(-8))
plt.xlabel(r'$\ell$')
plt.ylabel(r'$[\ell(\ell+1)/2\pi]  C_\ell^\mathrm{TT}$')
plt.plot(ll5,clTT5*ll5*(ll5+1)/2./pi,'r-', label='omega_k 0.5')
plt.plot(ll4,clTT4*ll4*(ll4+1)/2./pi,'b-',label='omega_bh 50%')
plt.plot(ll3,clTT3*ll3*(ll3+1)/2./pi,'g-',label='As 50%')
plt.plot(ll2,clTT2*ll2*(ll2+1)/2./pi,'y-',label='ns 50%')
plt.plot(ll1,clTT1*ll1*(ll1+1)/2./pi,'black',label='fiducial')
plt.legend()
```

1) omega_k 0.5 shifts the power spectrum to the right, so now the cosmological horizon is smaller. 
2) Omega_bh increase by 50% increases the amplitude of the temperature anisotropies compared to fiducial for low ls 
3) As increase by 50% increases the amplitude of the temperature anisotropies compared to fiducial for all ls
4) ns increase by 50% decreases the amplitude of the temperature anisotropies compared to fiducial for low ls
